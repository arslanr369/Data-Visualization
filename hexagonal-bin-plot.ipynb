```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Plot hexagonal bin plot
plt.hexbin(x, y, gridsize=20, cmap='Oranges')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hexagonal Bin Plot')

# Show plot
plt.show()
```
