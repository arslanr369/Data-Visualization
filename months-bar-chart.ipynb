```python
import matplotlib.pyplot as plt

# Data
labels = ['January', 'February', 'March', 'April', 'May', 'June']
values = [10, 12, 8, 15, 9, 11]
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']

# Plot
plt.bar(labels, values, color=colors)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')

# Show plot
plt.show()
```
