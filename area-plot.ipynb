```python
import matplotlib.pyplot as plt
import numpy as np

# Generate fake population data for a city over time
years = range(2000, 2020)
population = [10_000, 12_000, 14_000, 17_000, 20_000, 25_000, 30_000, 35_000, 40_000, 45_000, 
              50_000, 55_000, 60_000, 65_000, 70_000, 75_000, 80_000, 85_000, 90_000, 100_000]

# Create stacked area plot
plt.stackplot(years, population, labels=['Population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth of a City')

# Add legend and show plot
plt.legend(loc='upper left')
plt.show()
```
